```python
# RLE Compression Function
def run_length_encode(text):
    encoding = []
    i = 0
    while i < len(text):
        count = 1
        while i + 1 < len(text) and text[i] == text[i + 1]:
            i += 1
            count += 1
        encoding.append(f"{text[i]}{count}")
        i += 1
    return ''.join(encoding)
```

```python
# RLE Decompression Function
def run_length_decode(encoded_text):
    decoded_text = []
    i = 0
    while i < len(encoded_text):
        char = encoded_text[i]
        count = int(encoded_text[i+1])
        decoded_text.append(char * count)
        i += 2
    return ''.join(decoded_text)
```

```python
# Read input from a file
with open("Compression and decompression techniques.txt", "r") as file:
    text = file.read()

# Compress the text using RLE
encoded_rle = run_length_encode(text)
print("RLE Encoded Text:", encoded_rle)

# Save the encoded text to an output file
with open("encoded_output_rle.txt", "w") as file:
    file.write(encoded_rle)
```

```python
# Decompress the RLE encoded text
decoded_rle = run_length_decode(encoded_rle)
print("RLE Decoded Text:", decoded_rle)

# Optionally, save the decoded text to verify correctness
with open("decoded_output_rle.txt", "w") as file:
    file.write(decoded_rle)
```

```python
import heapq
from collections import defaultdict, Counter

# Step 1: Huffman Encoding
def huffman_encode(text):
    frequency = Counter(text)
    heap = [[weight, [char, ""]] for char, weight in frequency.items()]
    heapq.heapify(heap)

    while len(heap) > 1:
        lo = heapq.heappop(heap)
        hi = heapq.heappop(heap)
        for pair in lo[1:]:
            pair[1] = '0' + pair[1]
        for pair in hi[1:]:
            pair[1] = '1' + pair[1]
        heapq.heappush(heap, [lo[0] + hi[0]] + lo[1:] + hi[1:])

    huffman_codes = sorted(heap[0][1:], key=lambda p: (len(p[-1]), p))
    huffman_dict = {char: code for char, code in huffman_codes}

    encoded_text = ''.join(huffman_dict[char] for char in text)
    return huffman_dict, encoded_text
```

```python
# Step 2: Huffman Decoding
def huffman_decode(encoded_text, huffman_dict):
    reverse_dict = {code: char for char, code in huffman_dict.items()}
    decoded_text = []
    code = ""

    for bit in encoded_text:
        code += bit
        if code in reverse_dict:
            decoded_text.append(reverse_dict[code])
            code = ""

    return ''.join(decoded_text)
```

```python
# Read input from file
with open("Compression and decompression techniques.txt", "r") as file:
    text = file.read()

# Compress the text using Huffman Encoding
huffman_dict, encoded_huffman = huffman_encode(text)
print("Huffman Encoded Text:", encoded_huffman)

# Save the encoded Huffman text to an output file
with open("encoded_output_huffman.txt", "w") as file:
    file.write(encoded_huffman)
```

```python
# Decompress the Huffman encoded text
decoded_huffman = huffman_decode(encoded_huffman, huffman_dict)
print("Huffman Decoded Text:", decoded_huffman)

# Save the decoded text to verify correctness
with open("decoded_output_huffman.txt", "w") as file:
    file.write(decoded_huffman)
```
